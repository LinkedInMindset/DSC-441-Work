### c.) Regardless of the original ranges of the variables, normalization techniques transformthe data into new ranges that allow to compare and use variables on the same scales. What are thevalues ranges of the following normalization methods (for this data set and in general)? Explain and backup your answer.

#### i.)

```python
#Min Max Normalization
data = [[26, 10.5], [26, 30.5], [29, 8.8], [29, 20.8], [40, 32.4], [45, 26.9], [50, 30.4], [55, 30.2], [60, 33.2], [55, 36.6], [45, 44.5], [60, 30.8], [55, 35.4], [61, 33.2], [62, 36.1], [62, 36.1], [63, 37.9], [75, 43.2], [66, 37.7]]
df = pd.DataFrame(data, columns = ['Age','%Fat'])

#Min Max normalize Age
x = df['Age'][:]

x[:] = (x - x.min()) / (x.max() - x.min()) * (1-0) + 0

print('\n The Min-Max Normalization approach returns the following array for Age:\n')
print(x)

#Min Max normalize Fat
x = df['%Fat'][:]

x[:] = ((x - x.min()) / (x.max() - x.min())) * (1 - 0) + 0

print('\n The Min-Max Normalization approach returns the following array for %Fat:\n')
print(x)
```

The max is 1 and the min is 0 for both variables. This represents the range in general for every set of data because the max and min are relative to the dataset you're testing with.

The equation is :

$ \frac{(x_i - min(x))}{( max(x) - min(x))} * (new.max - new.min) + new.min$

#### ii.)

```python
#zscores
data = [[26, 10.5], [26, 30.5], [29, 8.8], [29, 20.8], [40, 32.4], [45, 26.9], [50, 30.4], [55, 30.2], [60, 33.2], [55, 36.6], [45, 44.5], [60, 30.8], [55, 35.4], [61, 33.2], [62, 36.1], [62, 36.1], [63, 37.9], [75, 43.2], [66, 37.7]]
df = pd.DataFrame(data, columns = ['Age','%Fat'])

age_zscore = stats.zscore(df['Age'])
print('The zscore is:')
print(age_zscore)

min_age = min(age_zscore)
print('\nThe minimum value computed in the zscore is:')
print(min_age)

max_age = max(age_zscore)
print('\nThe maximum value computed by the zscore is:')
print(max_age)

range_age = max_age - min_age
print('\nThe total range of the zscore is:')
print(range_age)
```

The zscore in this data ranges safely between -2 and 2 but z-score normalization doesn't always produce data with the same scale as before. 

Z-score doesn't necessarily have a fixed range. Because the equation is represented as :

$ \frac{(x_i - mu)}{(std)} $

a value, x, of the dataset can in theory get infinitly close to mu without touching mu, making the range indeterminable

### This problem is an example of data preprocessing needed in a data mining process.   Suppose a group of 12 sales price records has been sorted as follows: 8, 13, 14, 15, 17, 37, 55, 60, 77, 95, 208, 218.

```python
#2 BINS
sales = [8, 13, 14, 15, 17, 37, 55, 60, 77, 95, 208, 218]
df = pd.DataFrame(sales)
sns.lineplot(data = df)
```

```python
#Equal Depth

bin1 = sales[0:4]
bin2 = sales[4:8]
bin3 = sales[8:12]

def smoothMean(x):
    
    '''Take an array and replace values inside the array with the array's average'''
    
    x1 = []
    for i in x:
        x1.append(statistics.mean(x))
        
    return x1

bin1 = smoothMean(bin1)

bin2 = smoothMean(bin2)

bin3 = smoothMean(bin3)

plot_ED = bin1 + bin2 + bin3

df = pd.DataFrame(plot_ED)
sns.lineplot(data = df)
```

###### This in hindsight does a good job at predicting jumps. Note I say hindsight because we don't know the data is going to spike up until the data is alread collected but using this method could prove useful in financial markets where you are trying to follow trends while mitigating noise.

```python
#equal-width

bin1 = sales[0:6]
bin2 = sales[6:10]
bin3 = []
bin4 = sales[10:13]

bin1 = smoothMean(bin1)
bin2 = smoothMean(bin2)
bin3 = smoothMean(bin3)
bin4 = smoothMean(bin4)

df = pd.DataFrame(bin1 + bin2 + bin3 + bin4)

sns.lineplot(data = df)
```

###### This graph is in a way the opposite of the previous binning method. While looking at the previous plot in hindsight we saw that it predicted sharp jumps before they were done happening, this one starts late. In a way this is more reactive while the other is proactive. 

###### Depending on what you're looking for this could be helpful. It tells us that most of the values are bellow 100, which they are, while the previous graph makes it look like there is an even distribution of values across the dataset.

## b.) For the data in Figure 2 with three variables (X, Y, and Z) and two classes (I and II): which variable you would choose to classify the data? Show all the steps of your calculations and interpret your answer.

```python
from math import log2
df = pd.DataFrame([(1, 1, 1, 'A'), (1, 1, 1, 'A'), (0, 0, 1, 'B'), (1, 0, 0, 'B')],
                  columns=['X', 'Y', 'Z', 'Class'])
df
```

```python
#Total Entropy
Total_Entropy = -(1/2) * log2(1/2) - (1/2) * log2(1/2)

print('The total entropy is {}'.format(Total_Entropy))


#X_Gain
X_Entropy = (1/4) * (-(1) * log2(1)) + (3/4) * (-(2/3) * log2(2/3) - (1/3) * log2(1/3))
print('\n X entropy is {}'.format(X_Entropy))

X_Gain = Total_Entropy - X_Entropy
print('\n X Gain is {}'.format(X_Gain))


#Y_Gain
Y_Entropy = (1/2) * (-(1) * log2(1)) + (1/2) * (-(1) * log2(1))
print('\n Y entropy is {}'.format(abs(Y_Entropy)))

Y_Gain = Total_Entropy - Y_Entropy
print('\n Y Gain is {}'.format(Y_Gain))


#Z Gain
Z_Entropy = (1/4) * (-(1) * log2(1)) + (3/4) * (-((2/3) * log2(2/3)) - ((1/3) * log2(1/3)))
print('\n Z entropy is {}'.format(abs(Z_Entropy)))

Z_Gain = Total_Entropy - Z_Entropy
print('\n Z Gain is {}'.format(Z_Gain))
```

I would use variable Y. Variable Y appears to have the highest degree of influence when it comes to selecting the answer.

It not only completely mirrors the class assignment, but Y also has a information gain of 1 which is the highest of the 3 attributes.
